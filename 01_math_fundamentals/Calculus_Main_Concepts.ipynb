## 1. Number Theory
Number Theory studies properties of integers, including divisibility, prime numbers, and modular arithmetic.

**Key Concepts**:
- **Divisibility**: An integer \(a\) divides \(b\) if there exists an integer \(k\) such that \(b = a \cdot k\).
- **Prime Numbers**: Numbers greater than 1 divisible only by 1 and themselves.
- **GCD and LCM**: Greatest Common Divisor (GCD) and Least Common Multiple (LCM) are computed using the Euclidean algorithm.

**Example**: Find the GCD of 48 and 18.

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

print(f'GCD of 48 and 18 is {gcd(48, 18)}')
```

## 2. Order of Operations
The order of operations (PEMDAS) dictates the sequence for evaluating expressions: Parentheses, Exponents, Multiplication and Division (left to right), Addition and Subtraction (left to right).

**Example**: Evaluate \(2 + 3 \cdot 4^2\).

```python
result = 2 + 3 * 4**2
print(f'2 + 3 * 4^2 = {result}')
```

## 3. Variables
Variables represent unknown or changing quantities, often denoted by letters like \(x\) or \(y\).

**Example**: Solve for \(x\) in \(2x + 3 = 7\).

```python
from sympy import symbols, solve

x = symbols('x')
equation = 2*x + 3 - 7
solution = solve(equation, x)
print(f'Solution for x: {solution[0]}')
```

## 4. Functions
A function \(f(x)\) maps an input \(x\) to an output. For example, \(f(x) = x^2 + 2x + 1\).

**Example**: Plot \(f(x) = x^2 + 2x + 1\).

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 2*x + 1

x = np.linspace(-5, 5, 100)
y = f(x)

plt.plot(x, y, label='f(x) = x^2 + 2x + 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function Plot')
plt.grid(True)
plt.legend()
plt.show()
```

## 5. Summations
Summations represent the sum of a sequence, e.g., \(\sum_{i=1}^n i = 1 + 2 + \dots + n\).

**Example**: Compute \(\sum_{i=1}^{10} i^2\).

```python
n = 10
sum_squares = sum(i**2 for i in range(1, n+1))
print(f'Sum of squares from 1 to {n} is {sum_squares}')
```

## 6. Exponents
Exponents represent repeated multiplication, e.g., \(a^b\). Rules include \(a^m \cdot a^n = a^{m+n}\) and \((a^m)^n = a^{mn}\).

**Example**: Simplify \(2^3 \cdot 2^4\).

```python
result = 2**3 * 2**4
print(f'2^3 * 2^4 = {result} (or 2^{3+4} = 2^7 = {2**7})')
```

## 7. Logarithms
Logarithms are the inverse of exponents: \(\log_b(a) = c\) means \(b^c = a\).

**Example**: Compute \(\log_2(8)\).

```python
from math import log2

result = log2(8)
print(f'log_2(8) = {result}')
```

## 8. Eulerâ€™s Number and Natural Logarithms

### Eulerâ€™s Number
Eulerâ€™s number \(e \approx 2.718\) is the base of natural logarithms, fundamental in growth and decay processes.

### Natural Logarithms
The natural logarithm \(\ln(x) = \log_e(x)\) is the logarithm with base \(e\).

**Example**: Plot \(y = e^x\) and \(y = \ln(x)\).

```python
x_exp = np.linspace(-2, 2, 100)
y_exp = np.exp(x_exp)
x_ln = np.linspace(0.1, 5, 100)
y_ln = np.log(x_ln)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x_exp, y_exp, label='y = e^x')
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_ln, y_ln, label='y = ln(x)', color='orange')
plt.title('Natural Logarithm')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## 9. Limits
A limit describes the value a function approaches as the input approaches a point: \(\lim_{x \to a} f(x)\).

**Example**: Compute \(\lim_{x \to 2} (x^2 + 1)\).

```python
from sympy import limit, symbols

x = symbols('x')
f = x**2 + 1
limit_value = limit(f, x, 2)
print(f'Limit as x approaches 2 of x^2 + 1 is {limit_value}')
```

## 10. Derivatives
The derivative \(f'(x)\) represents the rate of change of a function. For \(f(x) = x^2\), \(f'(x) = 2x\).

**Example**: Compute the derivative of \(f(x) = x^3 + 2x\).

```python
from sympy import diff

x = symbols('x')
f = x**3 + 2*x
derivative = diff(f, x)
print(f'Derivative of x^3 + 2x is {derivative}')
```

## 11. Partial Derivatives
For functions of multiple variables, partial derivatives measure the rate of change with respect to one variable. For \(f(x, y) = x^2 + y^2\), \(\frac{\partial f}{\partial x} = 2x\).

**Example**: Compute partial derivatives of \(f(x, y) = x^2y + y^3\).

```python
y = symbols('y')
f = x**2 * y + y**3
partial_x = diff(f, x)
partial_y = diff(f, y)
print(f'Partial derivative with respect to x: {partial_x}')
print(f'Partial derivative with respect to y: {partial_y}')
```

###Calculate Partial derivaives using SymPY

```python
from sympy import *
from sympy.plotting import plot3d
#declaring x and y
x,y=symbols('x y')
f=2*x**3 + 3*y**3
dx_f=diff(f,x)
dx_y=diff(f,y)
print(dx_f)
print(dx_y)
#plot the function
plot3d(f)
```

## 12. The Chain Rule
The chain rule states that for composite functions \(f(g(x))\), the derivative is \(f'(g(x)) \cdot g'(x)\).

**Example**: Differentiate \(f(x) = \sin(x^2)\).

```python
from sympy import sin

f = sin(x**2)
derivative = diff(f, x)
print(f'Derivative of sin(x^2) is {derivative}')
```

## 13. Integrals
Integrals compute the area under a curve or the antiderivative. For \(f(x) = x^2\), \(\int x^2 \, dx = \frac{x^3}{3} + C\).

**Example**: Compute \(\int (x^2 + 2x + 1) \, dx\).

```python
from sympy import integrate

f = x**2 + 2*x + 1
integral = integrate(f, x)
print(f'Integral of x^2 + 2x + 1 is {integral}')
```
